This is a notebook to simulate dice rolls for DnD.

```python
import matplotlib.pyplot as plt
import numpy as np

from modsim import *
```

```python
def roll_die_avg(n_sides, n_times):
    rolls = np.random.uniform(1, n_sides+1, n_times)
    
    fig = plt.hist(rolls, n_sides, align='left', rwidth=0.8)
    plt.xlabel('Rolled Value')
    plt.ylabel('Frequency')
    
    plt.show()
    
    
    
    return np.mean(rolls)
```

```python
roll_die_avg(20, 1000000)
```

The shape of the histogram is uniform since I am generating a uniform distribution. The mean is a little of since the expected value is 10.5 not 11.

```python
def roll_die_max(n_sides, m_times):
    rolls = np.random.uniform(1, n_sides+1, m_times)
    
    return np.max(rolls)
```

```python
results = []
for i in range(1000000):
    results.append(roll_die_max(20, 2))
    
fig = plt.hist(results, 20, align='left', rwidth=0.8)
plt.xlabel('Rolled Value')
plt.ylabel('Frequency')
plt.plot()
```

Yes, higher values since on the first roll if you get a one, there is a 1/20 chance that you will keep that value. For 2 there is a 2/20 chance and so on until 20 where there is a 20/20 chance. Thus taken any one roll, n, that chance that you keep that value is n/20.
